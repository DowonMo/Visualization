```python
# 색상구분을 위한 구동 방식 유형 변경

drv_dict = {'f':'전륜 구동',
            '4':'사륜 구동',
            'r':'후륜 구동'}


def reset_dict(x):
    return drv_dict[x]

# df['drv'].map(drv_dict).value_counts()
# df['drv'].apply(lambda x: drv_dict[x])

df['drv'] = df['drv'].apply(reset_dict)
```

### (3) Stripplot
- dodge: hue 매개변수 사용시 dodge 매개변수의 값이 True일 경우 범주형 데이터 축을 따라서 strips 데이터들을 분리
- jitter: True일 경우 데이터의 개수가 많을 때 겹치지 않도록 조정해 데이터 분포의 이해도 향상
- size: 마커의 크기 지정

```python
plt.figure()
ax = sns.stripplot(data = df,
              x = 'drv',
              y = 'hwy',
              palette = 'deep',
              size = 3,
              jitter = True)
ax.set_title('구동 방식에 따른 주행 마일수(stripplot)')
ax.set_xlabel('구동 방식')
ax.set_ylabel('주행 마일수')
```

```python
# hue 매개변수로 실린더 수 구별

plt.figure()
ax = sns.stripplot(data = df,
              x = 'drv',
              y = 'hwy',
              palette = 'deep',
              size = 3,
              jitter = True,
              hue = 'cyl')
ax.set_title('구동 방식에 따른 주행 마일수(stripplot)')
ax.set_xlabel('구동 방식')
ax.set_ylabel('주행 마일수')
plt.legend(loc = 1)
```

```python
# dodge 매개변수를 활용해 hue로 구별된 값들을 그래프상 분리

plt.figure()
ax = sns.stripplot(data = df,
              x = 'drv',
              y = 'hwy',
              palette = 'deep',
              size = 3,
              jitter = True,
              hue = 'cyl',
              dodge = True)
ax.set_title('구동 방식에 따른 주행 마일수(stripplot)')
ax.set_xlabel('구동 방식')
ax.set_ylabel('주행 마일수')
plt.legend(loc = 1)
```

### (4) swarmplot
- stripplot 함수와 유사하지만, 점들이 보정돼 겹치지 않음
- 데이터 양이 많을 경우 비효율적임

```python
plt.figure()
ax = sns.swarmplot(data = df,
              x = 'drv',
              y = 'hwy',
              palette = 'deep')
ax.set_title('Swarm Plot')
ax.set_xlabel('구동 방식')
ax.set_ylabel('주행 마일수')
```

```python
# hue 옵션으로 실린더 수 구별

plt.figure()
ax = sns.swarmplot(data = df,
              x = 'drv',
              y = 'hwy',
              palette = 'deep',
              hue = 'cyl')
ax.set_title('Swarm Plot')
ax.set_xlabel('구동 방식')
ax.set_ylabel('주행 마일수')
```

```python
# dodge 매개변수를 활용해 hue로 구별된 값들을 그래프상 분리

plt.figure()
ax = sns.swarmplot(data = df,
              x = 'drv',
              y = 'hwy',
              palette = 'deep',
              hue = 'cyl',
              dodge = True)
ax.set_title('Swarm Plot')
ax.set_xlabel('구동 방식')
ax.set_ylabel('주행 마일수')
```

```python
# stripplot과 상자 수염 그래프 혼합

plt.figure()
sns.boxplot(data = df,
            x = 'displ',
            y = 'drv',
            palette = 'deep')
ax = sns.stripplot(data = df,
            x = 'displ',
            y = 'drv',
            color = '0.4')

ax.set_title('Boxplot과 Stripplot 동시 사용')
ax.set_xlabel('엔진 크기')
ax.set_ylabel('')
```

```python
# violinplot과 swarmplot 그래프 혼합

plt.figure()
sns.violinplot(data = df,
            x = 'drv',
            y = 'hwy',
            palette = 'deep',
            inner = None)
ax = sns.swarmplot(data = df,
            x = 'drv',
            y = 'hwy',
            color = '0.9')

ax.set_title('violinplot과 swarmplot 동시 사용')
ax.set_xlabel('구동 방식')
ax.set_ylabel('엔진 크기')
```
